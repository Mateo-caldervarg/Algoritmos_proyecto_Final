# Canales de Potasio

Código de la ecuación alpha_n y gráfica. Falta encontrar el epsilon para la vecindad donde la ecuación se indetermina. 

```julia
epsilon= 0.000000000000001

function alpha_n(V)
    α_n = (0.01 * (10 .- V)) / (ℯ^((10 .- V) / 10) .- 1)

if abs(V-10) < epsilon
        return 0.1
    else 
        return α_n
    end 
end
```

```julia
V = -1:20

Y = alpha_n.(V)

plot(V, Y)
```

```julia
function beta_n(V) 
    x = 0.125*ℯ^(-V/80)
    return x
end
```

```julia
V = -150:150
α = alpha_n.(V)
β = beta_n.(V)

plot(V, α)
plot!(V, β)
```

```julia
#β_m
function beta_m(V)
    x = 4*ℯ^-(V/18)
    return x
end
```

```julia
#α_h
function alpha_h(V)
    x = 0.07*ℯ^(-V/20)
    return x
end
```

```julia
#β_h
function beta_h(V)
    x = 1/(ℯ^((30-V)/10)+1)
    return x
end
```

```julia
#τ_n
function tau_inf(alpha, beta)
    x = 1 ./(alpha .+ beta)
    return x
end
```

```julia
function tau_n(alpha_n, beta_n)
    x = 1 ./(alpha_n .+ beta_n)
    return x
end
```

```julia
#n_infinito
function n_inf(alpha, beta)
    x = alpha ./ (alpha .+ beta)
    return x
end
```

```julia
#n_infinito
function n_n(alpha_n, beta_n)
    x = alpha ./ (alpha_n .+ beta_n)
    return x
end
```

```julia
function euler(f, α, a, b, N)
    n1 = N + 1
    u = zeros(n1, 2) 
    h = (b - a)/N
    u[1,1] = a
    u[1,2] = α 
  
    for i in 2:n1
        u[1,1] = u[i-1, 2] + h * f(u[i-1], u[i-1,2])
        u[i,1] = a + (i-1) * h
    end
    return u
end 
```

```julia
GK(gK, n) = gK * (n ^ 4.0)
GNa(gNa, m) = gNa * (m ^ 3.0) * h
GL(gL) = gL

#definimos las ecuaciones diferenciales del modelo H. H.

minT = 0.0
δT = 0.01
maxT=50

T = minT:δT:maxT

function dv(Vm, GK, GNa, GL, Cm, EK, ENa, El, I, dt);
    ((I  - (GK * (v - EK)) - (GNa * (v - ENa)) - (GL * (v - El))) / Cm) * dt
end

dn(n, Vm, dt) = ((alpha_n(Vm) * (1.0 - n)) - (beta_n(Vm) * n)) * dt
dm(m, Vm, dt) = ((alpha_m(Vm) * (1.0 - m)) - (beta_m(Vm) * m)) * dt
dh(h, Vm, dt) = ((alpha_h(Vm) * (1.0 - h)) - (beta_h(Vm) * h)) * dt
```

```julia
tmin = 0.0
tmax = 1000.0
dt = 0.01
T = tmin:dt:tmax
 
# Parametros
gK = 35.0
gNa = 40.0
gL = 0.3
Cm = 1.0
EK = -77.0
ENa = 55.0
El = -65.0
```

```julia
#función de tasa de cambio de canales de K+
alpha_n(Vm) = (0.02 * (Vm - 25.0)) / (1.0 - exp((-1.0 * (Vm - 25.0)) / 9.0))
beta_n(Vm) = (-0.002 * (Vm - 25.0)) / (1.0 - exp((Vm - 25.0) / 9.0))
 
#función de tasa de cambio de canales de Na+
alpha_m(Vm) = (0.182*(Vm + 35.0)) / (1.0 - exp((-1.0 * (Vm + 35.0)) / 9.0))
beta_m(Vm) = (-0.124 * (Vm + 35.0)) / (1.0 - exp((Vm + 35.0) / 9.0))
 
alpha_h(Vm) = 0.25 * exp((-1.0 * (Vm + 90.0)) / 12.0)
beta_h(Vm) = (0.25 * exp((Vm + 62.0) / 6.0)) / exp((Vm + 90.0) / 12.0)
 
# n, m & h 
n_inf(Vm=0.0) = alpha_n(Vm) / (alpha_n(Vm) + beta_n(Vm))
m_inf(Vm=0.0) = alpha_m(Vm) / (alpha_m(Vm) + beta_m(Vm))
h_inf(Vm=0.0) = alpha_h(Vm) / (alpha_h(Vm) + beta_h(Vm))
```

```julia
# Conductancia
GK(gK, n) = gK * (n ^ 4.0)
GNa(gNa, m) = gNa * (m ^ 3.0) * h
GL(gL) = gL
```

```julia
# Ecuaciones diferenciales
function dv(Vm, GK, GNa, GL, Cm, EK, ENa, El, I, dt);
    ((I  - (GK * (v - EK)) - (GNa * (v - ENa)) - (GL * (v - El))) / Cm) * dt
end
dn(n, Vm, dt) = ((alpha_n(Vm) * (1.0 - n)) - (beta_n(Vm) * n)) * dt
dm(m, Vm, dt) = ((alpha_m(Vm) * (1.0 - m)) - (beta_m(Vm) * m)) * dt
dh(h, Vm, dt) = ((alpha_h(Vm) * (1.0 - h)) - (beta_h(Vm) * h)) * dt
I = T * 0.002
```

```julia
# Condiciones iniciales
v = -65
m = m_inf(v)
n = n_inf(v)
h = h_inf(v)
 
v_result = Array{Float64}(undef, length(T))
m_result = Array{Float64}(undef, length(T))
n_result = Array{Float64}(undef, length(T))
h_result = Array{Float64}(undef, length(T))
 
v_result[1] = v
m_result[1] = m
n_result[1] = n
h_result[1] = h
 
 
for t = 1:length(T)-1
    GKt = GK(gK, n)
    GNat = GNa(gNa, m)
    GLt = GL(gL)
 
    dvt = dv(v, GKt, GNat, GLt, Cm, EK, ENa, El, I[t], dt)
    dmt = dm(m, v, dt)
    dnt = dn(n, v, dt)
    dht = dh(h, v, dt)
 
    global v = v + dvt
    global m = m + dmt
    global n = n + dnt
    global h = h + dht
 
    v_result[t+1] = v
    m_result[t+1] = m
    n_result[t+1] = n
    h_result[t+1] = h
end
```

```julia
plot(T, v_result, xlabel="time (ms)", ylabel="voltage (mV)", legend=false)
```
