# Canales de Potasio

Código de la ecuación alpha_n y gráfica. Falta encontrar el epsilon para la vecindad donde la ecuación se indetermina. 

```julia
function alpha_n(V)
    x = (0.01 * (10 - V)) / (ℯ^((10-V) / 10) .- 1)
    return x
end
```

```julia
alpha_n(10)
```

```julia
V = -1:20

Y = alpha_n.(V)

plot(V, Y)
```

```julia
function beta_n(V) 
    x = 0.125*ℯ^(-V/80)
    return x
end
```

```julia
V = -150:150
α = alpha_n.(V)
β = beta_n.(V)

plot(V, α)
plot!(V, β)
```

# Canales de Sodio

```julia
#α_m
function alpha_m(V)
    x = (0.1*(25-V))/(ℯ^((25-V)/10)-1)
    return x
end
```

```julia
#β_m
function beta_m(V)
    x = 4*ℯ^-(V/18)
    return x
end
```

```julia
#α_h
function alpha_h(V)
    x = 0.07*ℯ^(-V/20)
    return x
end
```

```julia
#β_h
function beta_h(V)
    x = 1/(ℯ^((30-V)/10)+1)
    return x
end
```

```julia
#τ_n
function tau_inf(alpha, beta)
    x = 1/(alpha + beta)
    return x
end
```

```julia
#n_infinito
function n_inf(alpha, beta)
    x = alpha / (alpha + beta)
    return x
end
```

```julia
V = 20

α = alpha_n(V)
β = beta_n(V)

n_inf(α, β)
```

```julia
V = -100:100

n_inf_n = n_inf(alpha_n(V), beta_n(V))
tau_inf_n = tau_inf(alpha_n(V), beta_n(V))
    
n_inf_m = n_inf(alpha_m(V), beta_m(V))
tau_inf_m = tau_inf(alpha_m(V), beta_m(V))
    
n_inf_h = n_inf(alpha_h(V), beta_h(V))
tau_inf_h = tau_inf(alpha_h(V), beta_h(V))
```

# Simulación con método numérico de Euler

```julia
using Pkg
Pkg.add("DifferentialEquations")
```

```julia
using DifferentialEquations
```

```julia
function metodo_euler(f, α, a, b, N)
    
    n1 = N + 1
    u = zeros(n1, 2) 
    h = (b - a)/N
    u[1,1] = a
    u[1,2] = α 
  
    for i in 2:n1
        u[1,1] = u[i-1, 2] + h * f(u[i-1], u[i-1,2])
        u[i,1] = a + (i-1) * h
    end
    return u
end 
```
